```python
# print alphabet pattern
x=0
while  x<8:
    content=""
    y=0
    while y<x:
        z=0
        sum=0
        while z<y:
            sum+=6-z
            z+=1
        if(x+64+sum)<=(64+26):
            content+=" "+chr(x+64+sum)
        y+=1
    print(content)
    x+=1
            
```

```python
# input dictionary output list
def d_to_l(d):
    li=[]
    for k,v in d.items():
        if type(v)==dict:
            for i,j in v.items():
                li.append(i)
                li.append(j)
        elif type(k)==dict:
             for m,n in k.items():
                li.append(m)
                li.append(n)
        else:
            li.append(k)
            li.append(v)
    print(li)        
            
```

```python
d_to_l(d)
```
